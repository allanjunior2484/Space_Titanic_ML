No ano de 2912, a nave espacial Titanic, transportando cerca de 13.000 passageiros emigrantes entre sistemas solares, encontrou um destino trágico ao colidir com uma anomalia do espaço-tempo perto de Alfa Centauri. Este evento resultou na transferência de quase metade dos passageiros para uma dimensão alternativa, enquanto a nave permaneceu fisicamente intacta. Minha missão é resolver o mistério da nave espacial Titanic em 2912. Após colidir com uma anomalia do espaço-tempo perto de Alfa Centauri, quase metade dos passageiros foi transferida para uma dimensão alternativa, enquanto a nave permaneceu intacta.


Descrições de arquivos e campos de dados
train.csv - Registros pessoais de cerca de dois terços (~8.700) dos
passageiros, para serem usados ​​como dados de treinamento. **negrito**

**PassengerId**- Um ID único para cada passageiro. Cada Id assume a forma
gggg_pponde ggggindica o grupo com o qual o passageiro está viajando e ppé o seu número dentro do grupo. As pessoas de um grupo são frequentemente membros da família, mas nem sempre.  

**HomePlanet**-* O planeta de onde o passageiro partiu, normalmente o planeta de residência permanente.*

**CryoSleep**- *Indica se o passageiro optou por ser colocado em animação suspensa durante a viagem. Os passageiros em sono criogênico ficam confinados em suas cabines.*

**Cabin**- *O número da cabine onde o passageiro está hospedado. Assume a forma deck/num/side, sidepodendo ser tanto Ppara Bombordo como Spara Estibordo .*

**Destination**- *O planeta para onde o passageiro irá desembarcar.*

**Age**-* A idade do passageiro.*

**VIP**- *Se o passageiro pagou por serviço VIP especial durante a viagem.*

**RoomService**, **FoodCourt**, **ShoppingMall**, **Spa**, **VRDeck**- Valor *que o passageiro faturou em cada uma das muitas comodidades de luxo da nave espacial Titanic .*

**Name**- *O nome e sobrenome do passageiro.*

**Transported**-* Se o passageiro foi transportado para outra dimensão. Este é o alvo, a coluna que você está tentando prever.*

test.csv - Registros pessoais do terço restante (~4.300) dos passageiros, para serem usados ​​como dados de teste. Sua tarefa é prever o valor de Transportedpara os passageiros deste conjunto.

sample_submission.csv – Um arquivo de envio no formato correto.

** PassengerId**- ID de cada passageiro no conjunto de teste.

**Transported**- *O alvo. Para cada passageiro, preveja Trueou False.*

#feature engineering

```python
#train
train_df[['gggg', 'pp']] = train_df['PassengerId'].str.split('_', expand=True) #Foi nos passado o dado do PassengerId segue um padrão de gggg_pp isso foi comentado acima
#test
test_df[['gggg', 'pp']] = test_df['PassengerId'].str.split('_', expand=True)# sabendo disso vamos extrair esse pafrão e coloca-los em colunas distintas
```

```python
#train
uinique_gggg_train = train_df['gggg'].value_counts()#vamos estrir sua frequência e substituir seu valor por ela(frequência)
train_df['gggg_count'] = train_df['gggg'].map(uinique_gggg_train)
#test
uinique_gggg_test = test_df['gggg'].value_counts()
test_df['gggg_count'] = test_df['gggg'].map(uinique_gggg_test)
```

```python
#train
train_df[['deck', 'num', 'side']] = train_df['Cabin'].str.split('/', expand=True)#Temos també os dados da Cabin condensados vamos dividi-los pois assim é melhor para tirarmos a
#test
test_df[['deck', 'num', 'side']] = test_df['Cabin'].str.split('/', expand=True)#maior quantidade de unicidade e acharmos um padão válido
```

```python
#train
train_df['Surname'] = train_df['Name'].str.split(pat=' ', expand=True).iloc[:, -1]# vamos agora ver a frequência dos sobre-nomes que indica que o tripulante é da mesma familía
#test
test_df['Surname'] = test_df['Name'].str.split(pat=' ', expand=True).iloc[:, -1]
```

```python
#train
unique_surname_train = train_df['Surname'].value_counts()#extraimos a freqência
train_df['Surname_count'] =  train_df['Surname'].map(unique_surname_train)# e então substituimos criamos uma nova coluna e mapeamos com o valor da freqência
#test
unique_surname_test = test_df['Surname'].value_counts()
test_df['Surname_count'] =  test_df['Surname'].map(unique_surname_test)
```

```python
train_df['Expanse'] = train_df[['RoomService',	'FoodCourt',	'ShoppingMall',	'Spa',	'VRDeck']].sum(axis=1)#Aqui criamos uma coluna com o valor total isso foi fez com que eu obtivesse mais precisão no meu modelo
test_df['Expanse'] = test_df[['RoomService',	'FoodCourt',	'ShoppingMall',	'Spa',	'VRDeck']].sum(axis=1)
```

```python
#Vamos eliminar as colunas que já não faz tanto sentido manter
```

```python
#train
train_df.drop(columns=['Cabin','num','gggg','Name','Surname'], axis=1, inplace=True)#Descartamos as colunas condensadas
#test
test_df.drop(columns=['Cabin','num','gggg','Name','Surname'], axis=1, inplace=True)
```



Básicamente vamos atribuir os valores das média para os dados númericos não categóricos e o valor mais frequente para os categóricos

```python
#train
train_df['gggg_count'] = train_df['gggg_count'].fillna(train_df['gggg_count'].mode())
#test
test_df['gggg_count'] = test_df['gggg_count'].fillna(test_df['gggg_count'].mode())
```

```python
#train
train_df['Surname_count'] = train_df['Surname_count'].fillna(train_df['Surname_count'].mode()[0])
#test
test_df['Surname_count'] = test_df['Surname_count'].fillna(test_df['Surname_count'].mode()[0])
```

```python
#train
numeric_column_train = train_df.select_dtypes(include=['number']).columns
#test
numeric_column_test = test_df.select_dtypes(include=['number']).columns
numeric_column_train
```

```python
imputer = SimpleImputer(strategy='mean')
```

```python
#train
imputer.fit(train_df[numeric_column_train])
#test
imputer.fit(test_df[numeric_column_test])
```

```python
#train
train_df[numeric_column_train] = imputer.transform(train_df[numeric_column_train])
train_df[numeric_column_train] = train_df[numeric_column_train].round(1)
#test
test_df[numeric_column_test] = imputer.transform(test_df[numeric_column_test])
test_df[numeric_column_test] = test_df[numeric_column_test].round(1)
```

#Padrozização

Vamos começar a padronização de nossos dados para receber o modelo de KKN

O KNN (K-Nearest Neighbors) é um método de aprendizado de máquina que decide a classe ou o valor de um novo ponto de dados baseado nos seus vizinhos mais próximos, usando medidas de distância.

```python
#train
train_df[['CryoSleep','VIP']] = train_df[['CryoSleep','VIP',]].astype(int)#vamos transforma os dados bool em int para facilitar a integração do padronizador scaler mais a baixo
#test
test_df[['CryoSleep','VIP']] = test_df[['CryoSleep','VIP']].astype(int)
```

#OneHotEncoder

O OneHotEncoder converte variáveis categóricas em vetores binários, onde cada categoria é representada por uma coluna. É amplamente usado em aprendizado de máquina para processar dados categóricos em formatos que modelos possam entender diretamente

```python
ohe = OneHotEncoder()
```

```python
#train
train_objec = ohe.fit(train_df[['HomePlanet','Destination','deck','side']])
train_objec.transform(train_df[['HomePlanet','Destination','deck','side']]).toarray()
#test
test_objec = ohe.fit(test_df[['HomePlanet','Destination','deck','side']])
test_objec.transform(test_df[['HomePlanet','Destination','deck','side']]).toarray()
```

```python
#train
train_objec = pd.DataFrame(train_objec.transform(train_df[['HomePlanet','Destination','deck','side']]).toarray(),columns=ohe.get_feature_names_out())
#test
test_objec = pd.DataFrame(test_objec.transform(test_df[['HomePlanet','Destination','deck','side']]).toarray(),columns=ohe.get_feature_names_out())
```

```python
#train
train_df.drop(columns = ['HomePlanet','Destination','deck','side'],axis=1,inplace=True)
#test
test_df.drop(columns = ['HomePlanet','Destination','deck','side'],axis=1,inplace=True)
```

```python
#train
train_df = pd.concat([train_df,train_objec], axis=1)#aqui estamos integrando as colunas criadas pelo OHE no nosso df
#test
test_df = pd.concat([test_df,test_objec],axis=1)
```

```python
train_df.head(3)
```

```python
#train
X_train = train_df.drop(columns=['PassengerId','Transported','RoomService',	'FoodCourt',	'ShoppingMall',	'Spa',	'VRDeck','Expanse_Cut'],axis=1)#excluímos os valores que não serão uteis para a previsão dos dados
y_train = train_df['Transported']#e definimos as variáveis que servirá de base para previsão X_train,y_train para treino e X_test,y_test para o teste
#test
X_test = test_df.drop(columns=['PassengerId','RoomService',	'FoodCourt',	'ShoppingMall',	'Spa',	'VRDeck','Expanse_Cut'],axis=1)
y_test =y_test_df.drop('PassengerId',axis=1)
```

```python
scaler = StandardScaler()
scaler.fit_transform(X_train,X_test);# Padronizamos com o scaler os dados que entrarão no knn
#O StandardScaler é usado para padronizar os dados, ajustando-os para
# que tenham média zero e desvio padrão um, o que facilita o treinamento de modelos ao garantir que as variáveis estejam na mesma escala.
```

```python
knn = KNeighborsClassifier(n_neighbors=2)
```

```python
knn.fit(X_train,y_train);#encaixamos nossas variáveis no knn
```

```python
previsoes = knn.predict(X_test)
```
